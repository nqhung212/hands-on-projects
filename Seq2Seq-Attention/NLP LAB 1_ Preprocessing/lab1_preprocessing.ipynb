# Lab 1: NLP Preprocessing


```python
from pathlib import Path


def tokenize(sentence):
    return sentence.lower().split()
```

```python
class Vocab:
    def __init__(self):
        self.word2id = {"<PAD>": 0, "<BOS>": 1, "<EOS>": 2}
        self.id2word = {0: "<PAD>", 1: "<BOS>", 2: "<EOS>"}

    def add_word(self, word):
        if word not in self.word2id:
            idx = len(self.word2id)
            self.word2id[word] = idx
            self.id2word[idx] = word

    def build(self, tokenized_sentences):
        for tokens in tokenized_sentences:
            for token in tokens:
                self.add_word(token)

    def encode(self, tokens, add_bos=True, add_eos=True):
        ids = []
        if add_bos:
            ids.append(self.word2id["<BOS>"])
        ids.extend(self.word2id[token] for token in tokens)
        if add_eos:
            ids.append(self.word2id["<EOS>"])
        return ids

    def __len__(self):
        return len(self.word2id)
```

## 1) Tokenization

```python
DATA_DIR = Path("data")

if not DATA_DIR.exists():
    raise FileNotFoundError("Không có data")


def read_lines(file_name):
    file_path = DATA_DIR / file_name
    with open(file_path, encoding="utf-8") as file:
        return [line.strip() for line in file if line.strip()]


train_en = read_lines("train.en")
train_vi = read_lines("train.vi")
dev_en = read_lines("dev.en")
dev_vi = read_lines("dev.vi")

tokenized_train_en = [tokenize(s) for s in train_en]
tokenized_train_vi = [tokenize(s) for s in train_vi]
tokenized_dev_en = [tokenize(s) for s in dev_en]
tokenized_dev_vi = [tokenize(s) for s in dev_vi]

print(f"train.en sentences: {len(train_en)}")
print(f"train.vi sentences: {len(train_vi)}")
print(f"dev.en sentences: {len(dev_en)}")
print(f"dev.vi sentences: {len(dev_vi)}")
```

## 2) Vocabulary Construction

```python
src_vocab = Vocab()
tgt_vocab = Vocab()

src_vocab.build(tokenized_train_en)
tgt_vocab.build(tokenized_train_vi)

print(f"English vocab size: {len(src_vocab)}")
print(f"Vietnamese vocab size: {len(tgt_vocab)}")
print("First 6 EN vocab items:", list(src_vocab.word2id.items())[:6])
```

## 3) Sentence Encoding

```python
sentence_en = dev_en[0]
tokens_en = tokenize(sentence_en)
encoded_en = src_vocab.encode(tokens_en, add_bos=True, add_eos=True)

sentence_vi = dev_vi[0]
tokens_vi = tokenize(sentence_vi)
encoded_vi = tgt_vocab.encode(tokens_vi, add_bos=True, add_eos=True)

print("Sentence EN:")
print(sentence_en)
print("Encoded EN:")
print(encoded_en)
print("Sentence VI:")
print(sentence_vi)
print("Encoded VI:")
print(encoded_vi)
```
